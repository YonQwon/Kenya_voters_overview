```python
#function to return the number of volters per County 
def cntcount(name):
    mask=df["County Name"]==name
    dt=df[mask]
    nor=dt["Registered Voters"].sum()
    return print(f"The nummber of regestered volters in {name} is {nor}.")
```

```python
#function to return a DataFrame of the County
def countydf (cname):
    dk=df[df["County Name"]==cname]
    return dk
   
```

```python
#function to return number of volters per constituency
def ctcount(ctname):
    mask=df["Const."]==ctname
    dct=df[mask]
    cnt=dct["Registered Voters"].sum()
    return cnt
```

```python
#function to return bar plot of 
def plot(cname):
     dk=df[df["County Name"]==cname]
     data=dk.groupby("Const.")["Registered Voters"].sum().sort_values(ascending=False)
     fig=px.bar(data_frame=data,
                color=data,
     title="Number of regestered votes in cname' COUNTY PER Constituencies")
     return fig.update_layout(xaxis_title="constituencies")
 
    
```

```python
#function to return number of volters per constituency
def numpercounty(cname): 
    dg=df[df["County Name"]==cname]
    dv=dg.groupby("Const.")["Registered Voters"].sum().sort_values(ascending=False)
    return dv


```
